```
import numpy as np
from matplotlib import pyplot as plt
# Image management
from skimage import io # Input/output
from skimage import transform # Geometric transformations (e.g. resize)
from skimage import filters # Local processing
from skimage import color # Change of color space
```

#Immagine scelta

```
im=io.imread('images/sea.jpg')
plt.imshow(im)
```

Visualizzazione della decomposizione della foto nei tre canali R,G,B: prima in scala di grigi e poi con le colormap specifiche per i tre canali

```
plt.subplot(2,2,1);plt.imshow(im[:,:,0], cmap='gray'); plt.title("Canale Rosso")
plt.subplot(2,2,2);plt.imshow(im[:,:,1], cmap='gray'); plt.title("Canale Verde")
plt.tight_layout()
plt.subplot(2,2,3);plt.imshow(im[:,:,2], cmap='gray'); plt.title("Canale Blu")
plt.subplot(2,2,4);plt.imshow(im); plt.title("Foto originale")
```

```
plt.subplot(2,2,1);plt.imshow(im[:,:,0], cmap='Reds'); plt.title("Canale Rosso")
plt.subplot(2,2,2);plt.imshow(im[:,:,1], cmap='Greens'); plt.title("Canale Verde")
plt.tight_layout()
plt.subplot(2,2,3);plt.imshow(im[:,:,2], cmap='Blues'); plt.title("Canale Blu")
plt.subplot(2,2,4);plt.imshow(im); plt.title("Foto originale")
```

#Conversione dell'immagine sea.jpeg in scala YcbCr

```
im_ycbcr=color.rgb2ycbcr(im)
plt.subplot(2,2,1);plt.imshow(im_ycbcr[:,:,0], cmap='gray'); plt.title("Y")
plt.subplot(2,2,2);plt.imshow(im_ycbcr[:,:,1], cmap='gray'); plt.title("Cb")
plt.tight_layout()
plt.subplot(2,2,3);plt.imshow(im_ycbcr[:,:,2], cmap='gray'); plt.title("Cr")
plt.subplot(2,2,4);plt.imshow(im); plt.title("Foto originale")
```

#Sfocatura dei canali Cb,Cr con filtro gaussiano

Sfocatura del canale Cb con un filtro gaussiano

```
plt.imshow(im_ycbcr[:,:,1], cmap='gray'); plt.title("Cb Originale")
```

```
im_cb=filters.gaussian(im_ycbcr[:,:,1],sigma=(5,3))
plt.imshow(im_cb, cmap='gray'); plt.title("Cb Filtrato")
```

Sfocatura del canale Cr con un filtro gaussiano

```
im_cr=filters.gaussian(im_ycbcr[:,:,2],sigma=(10,10))
plt.imshow(im_cr, cmap='gray'); plt.title("Cr Filtrato")
```

```
plt.imshow(im_ycbcr[:,:,2], cmap='gray'); plt.title("Cr Originale")
```

#Converto immagine filtrata nei canali Cb,Cr in RGB

```
im_modified_yCBCR=np.dstack((im_ycbcr[:,:,0], im_cb, im_cr))
im_modified_CBCR=color.ycbcr2rgb(im_modified_yCBCR)
plt.imshow(im_modified_CBCR); plt.title("Foto Filtrata")
```

```
plt.imshow(im); plt.title("Foto Originale")
```

L'immagine filtrata, come previsto,  ancora molto simile all'originale.
